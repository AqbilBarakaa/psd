```python
import os
import shutil
import pandas as pd
import numpy as np

print("MULAI PROSES SPLITTING & KOMPRESI DATASET...")

SOURCE_FILE = 'Strawberry_TEST.txt' 
BASE_OUTPUT_DIR = 'Strawberry_TEST_Per_Sample'

if os.path.exists(BASE_OUTPUT_DIR):
    shutil.rmtree(BASE_OUTPUT_DIR)

os.makedirs(f"{BASE_OUTPUT_DIR}/Authentic_Asli", exist_ok=True)
os.makedirs(f"{BASE_OUTPUT_DIR}/Adulterated_Palsu", exist_ok=True)

print(f"Folder output dibuat di: {BASE_OUTPUT_DIR}/")

try:
    df_split = pd.read_csv(SOURCE_FILE, sep=r"\s+", header=None)
    
    count_asli = 0
    count_palsu = 0
    
    print(f"Sedang memproses {len(df_split)} sampel...")
    
    for index, row in df_split.iterrows():
        label = int(row[0])
        features = row[1:].values
        
        if label == 1:
            count_asli += 1
            filepath = f"{BASE_OUTPUT_DIR}/Authentic_Asli/sample_asli_{count_asli}.txt"
        else:
            count_palsu += 1 
            filepath = f"{BASE_OUTPUT_DIR}/Adulterated_Palsu/sample_palsu_{count_palsu}.txt"
            
        np.savetxt(filepath, features.reshape(1, -1), fmt='%.6f', delimiter=' ')

    print(f"Berhasil memecah file:")
    print(f"   - {count_asli} file di folder Authentic_Asli")
    print(f"   - {count_palsu} file di folder Adulterated_Palsu")

    print("\nSedang mengompres folder menjadi .zip...")
    
    shutil.make_archive('Strawberry_TEST_Per_Sample', 'zip', BASE_OUTPUT_DIR)
    
    print(f"SELESAI! File zip siap diunduh: 'Strawberry_TEST_Per_Sample.zip'")

except FileNotFoundError:
    print(f"Error: File sumber '{SOURCE_FILE}' tidak ditemukan. Cek path dataset Anda.")
except Exception as e:
    print(f"Error tidak terduga: {e}")
```
