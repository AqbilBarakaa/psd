### Mengumpulkan Data Awal

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Memuat dataset dari file CSV
try:
    df = pd.read_csv('fetal_health.csv')
    print("Dataset berhasil dimuat!")
    print(f"Bentuk data: {df.shape[0]} baris dan {df.shape[1]} kolom.")
except FileNotFoundError:
    print("File 'fetal_health.csv' tidak ditemukan. Pastikan file berada di direktori yang sama.")

# Menampilkan 5 baris pertama dari data
df.head()
```

### Mendeskripsikan Data

```python
# Menampilkan informasi umum (tipe data dan nilai non-null)
print("Informasi Umum Dataset:")
df.info()

# Menampilkan statistik deskriptif
print("\nStatistik Deskriptif:")
df.describe()
```

### Eksplorasi Data / Menyelidiki Data

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribusi target
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='fetal_health', data=df)
plt.title('Distribusi Kelas Kesehatan Janin', fontsize=16)
plt.xticks([0, 1, 2], ['1: Normal', '2: Suspect', '3: Pathological'])
plt.show()

# Korelasi antar fitur
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap="coolwarm", center=0)
plt.title("Heatmap Korelasi antar Fitur")
plt.show()

# Boxplot salah satu fitur penting
plt.figure(figsize=(12, 7))
sns.boxplot(x='fetal_health', y='prolongued_decelerations', data=df)
plt.title('Hubungan antara Deselerasi Lama dan Kesehatan Janin', fontsize=16)
plt.xticks([0, 1, 2], ['1: Normal', '2: Suspect', '3: Pathological'])
plt.show()
```

### Verifikasi Kualitas Data

```python
# Menghitung jumlah nilai null di setiap kolom
missing_values = df.isnull().sum()
print("Jumlah nilai yang hilang per kolom:")
print(missing_values)
```

```python
# Menghitung jumlah baris duplikat
duplicate_rows = df.duplicated().sum()
print(f"\nJumlah baris data duplikat: {duplicate_rows}")
```
