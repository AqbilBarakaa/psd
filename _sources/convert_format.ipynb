convert audio buka

```python
import os
from pydub import AudioSegment

folder = r"C:\Semester5\psd\psd\uji\buka"
files = [f for f in os.listdir(folder) if f.endswith(".waptt")]

for idx, fname in enumerate(sorted(files)):
    src_path = os.path.join(folder, fname)
    # dst_path = os.path.join(folder, f"buka_uji_{idx}.wav")
    dst_path = os.path.join(folder, f"buka_uji_5.wav")
    # Konversi ke wav (ganti 'AudioSegment.from_file' jika format .waptt tidak dikenali, misal .mp3, .ogg, dll)
    audio = AudioSegment.from_file(src_path)
    audio.export(dst_path, format="wav")
    print(f"Converted: {src_path} -> {dst_path}")
```

untuk pernomoran lanjut dengan beda file directory

```python
import os
from pydub import AudioSegment

# Folder sebelumnya
folder1 = r"C:\Semester5\psd\psd\dataset\aqbil\tutup"
files1 = [f for f in os.listdir(folder1) if f.endswith(".wav") and f.startswith("tutup_")]
start_idx = len(files1)  # Penomoran lanjutan

# Folder baru
folder2 = r"C:\Semester5\psd\psd\dataset\reni\tutup"
files2 = [f for f in os.listdir(folder2) if f.endswith(".waptt")]

for idx, fname in enumerate(sorted(files2)):
    src_path = os.path.join(folder2, fname)
    dst_path = os.path.join(folder2, f"tutup_{start_idx + idx}.wav")
    audio = AudioSegment.from_file(src_path)
    audio.export(dst_path, format="wav")
    print(f"Converted: {src_path} -> {dst_path}")
```
